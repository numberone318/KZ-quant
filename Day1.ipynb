```python
# Day1 + 20   CFA 41
import akshare as ak
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimHei']   # 
plt.rcParams['axes.unicode_minus'] = False

# 12020
df = ak.stock_zh_a_hist(symbol="601939", start_date="20200101", adjust="qfq")

# 220
df['MA20'] = df[''].rolling(window=20).mean()

# 3
plt.figure(figsize=(14,7))
plt.plot(df[''], df[''], label='', linewidth=1.2)
plt.plot(df[''], df['MA20'], label='20', color='orange')
plt.title('(601939) 2020-2025 by KZ', fontsize=18)
plt.xlabel(''); plt.ylabel('()')
plt.legend(); plt.grid(); plt.tight_layout()
plt.show()

# 4
df.to_csv('C:/Users/Administrator/Desktop/.csv', 
          index=False, encoding='utf-8-sig')
print("Day1")
```
